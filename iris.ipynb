```python
#IMPORTAMOS PRIMERO LAS LIBRERIAS PARA EL PLOTEO, 
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
#IMPORTAMOS NUMPY PARA LAS OPERACIONES MATEMATICAS EN GENERAL
import numpy as np
#IMPORTAMOS PANDAS PARA LO QUE ES EL CARGADO DE EL DATASET
import pandas as pd
#IMPORTAMOS SKLEARN PARA APLICAR LO QUE ES LA UTILIZACION DE EL MODELO PARA LA CLASIFICIACION Y CLUSTERIZACION
from sklearn import neighbors, datasets
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
#LO QUE SE ESTA REALIZANDO ES EL CARGADO DE EL DATASET Y LA CUAL SE ESTAN MOSTRANDO TODOS LOS DATOS YA CARGADOS
iris = datasets.load_iris()
X = iris.data
Y = iris.target
print(X)
```

```python

#AQUI MOSTRAMOS LO  QUE ES EL TARGET(LOS IDENTIFICADORES DE LOS TIPOS DE FLORES) 
flowers = iris.target_names
print(flowers)
#MOSTRAMOS LO QUE ES EN QUE ES LO QUE SE ESTA MIDIENDO Y EN MEDIDA 
features = iris.feature_names
print(features)
```

```python
#LO QUE SE REALIZA ES UN PLOT DE LOS LAS ENTRADAS QUE SERIA : 
#LARGO DE SEPALO, ANCHO DE SEPALO, LARGO DE PETALO Y ANCHO DE PETALO
#EN ESTE CASO SE HARA UN PLOT DE LARGO DE PETALO Y LARGO DE SEPALO
plt.scatter(X[:, 0], X[:, 2], c=Y, cmap=plt.cm.Set1,
            edgecolor='w')
plt.title('Iris')
plt.xlabel('SEPALO LARGO')
plt.ylabel('PETALO LARGO')
```

```python
#SE ESTA REALIZANDO UN GRAFICO CAPTURANDO TODOS LOS CAMPO Y ESTA GRAFICANDO
#SEGUN AL TIPO DE YA SEA LARGO, ANCHO, DE PETALO Y CEPALO
df = pd.DataFrame(X, columns=features)
df.plot(figsize=(11, 11))
```

```python
#AHORA SE APLICARA LO QUE ES EL KNN O EL VECINO MAS PROXIMO
#Y LO QUE SE ESTA APLICANDO ES KNeighborsClassifier
model = neighbors.KNeighborsClassifier(n_neighbors=5)
#Y SE ESTA REALIZANDO EL ENTRENAMIENTO CON LA FUNCION FIT, CON EL DATASET COMO ENTRADA
model.fit(X, Y)
```

```python
#AHORA LOQUE SE ESTA HACEDO ES LA PREDICCION Y ESTO SE ESTA BASADOCE EN
#UN MODELO QUE EN ESTE CASO SE UNA RED NEURONAL DE LA CAPA DE ENTRADA QUE TENDRA 3 NEURONAS 
#LA CAPA OCULTA TENDRA DOS CAPAS  EN UNA CONTENDRA 5 Y EN LA OTRA 4 NEURONAS Y LA CAPA DE CALIDA DE 2 NEURONAS
y_predicted = model.predict([[3, 5, 4, 2]])
print("Predicted Y :: {}".format(y_predicted))
print("Target flower :: {}".format(iris.target_names[y_predicted]))
print("Probability for each flower :: {}".format(model.predict_proba([[3, 5, 4, 2]])))
```

```python
# AHORA LO QEU REALIZAMOS ESLADIVISIO DE LOS DATOS EN DATOS DE ENTRENAMEINTO Y DATOS DE PRUEBA

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1/3)

# SE USARA EL KNN EN EL DATASER DE IRIS

model = neighbors.KNeighborsClassifier(n_neighbors=3)

# SE ESTA REALIZANDO EL ENTRENAMIENTO CON LA PORCION DE DATOS DE ENTRENAMIENTO

model.fit(X_train, Y_train)

# SE ESTA REALIZANDO LA PREDICCION CON LOS DATOS DE PRUEBA

Y_pred = model.predict(X_test)

# AHORA SE ESTA CALCULANDO EL ACURACY QUE SN LOS SOBRE LOS DATOS DE PREDICCION Y DATOS DE PRUEBA

acc = np.sum(Y_pred == Y_test) /  len(Y_test)
print("acurracy:")
acc * 100
```

```python
# AHPRA LOQEU SE REALIZARA ES LA VALIDACION CRUZADA YA QUE ESTE METODO ES MUCHO MEJOR
# A LA HORA DE FRACCIONAR LO QUE SON LA PORCION DE DATASET DE ENTRENAMIENTO Y DE PRUEBA

# SEESTA ONSIDERANDO QUE SE REALIZARA 
klist = list(range(1,50, 2))

# empty list that will hold cv scores
cv_scores = []

# SE ESTA CONSIDERANDO UNA VALIDACION CRUZADA DE 10, ESTO INDICA QUE PARTICIONARA
# 10 VECES EL DATA SET EN 10 Y LUEGO DE ESA FORMA POCO A POCO CALCULARA EL ACURACY 
# DE CADA FOULT Y LUEGO CON CADA RESULTADO SACARA UN PROMEDIO.
for k in klist:
    knn = neighbors.KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())

# CAMBIANDO EL ERROR
errors = 1- np.array(cv_scores)

# DETERMINAND EL LEJOR VALOR DE K
optimal_k = klist[np.argmin(errors)]
print("EN NUMERO OPTIMO DE VECINO ES".format(optimal_k))

# MOSTRANDO  LA EVOLUCION DEL ERROR CONFORME A LAS ITERACIONES QUE HACE EN ESTE CASO 50 ITERACIONES
plt.plot(klist, errors)
plt.xlabel('NUMEROS DE VECINOS DE K')
plt.ylabel('EVOLUCION DE ERROR')
plt.show()
```
